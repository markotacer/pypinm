# Enakomerno asovno vzorenje in frekvenno prekrivanje

Pri procesiranju merjenih podatkov imamo opravka z diskretno vzorenimi signali. Pomembno se je zavedati lastnosti procesa diskretizacije sicer zveznih asovnih signalov, katerih osnove si bomo pogledali na tej vaji.

## Diskretizacija signalov

Diskretizacija je postopek, pri katerem v asu zvezno sprmenljivko pretvorimo v diskretne vrednosti, primerne za predstavitev v raunalnikem pomnilniku. Sestavljata jo dva loena procesa, **vzorenje** in **kvantizacija**.

### Enakomerno asovno vzorenje

Vzorenje je postopek diskretizacije *neodvisne spremenljivke* (asa) zveznega procesa. 

Glavni parameter asovnega vzorenja je **interval (perioda) vzorenja**, $\Delta t$, ki definira **frekvenco vzorenja**, $f_s = 1 / \Delta t$.

Zvezni asovni signal tako vzorimo pri diskretnih vrednostih asa $n \, \Delta t$.

:::{note}
Spomnimo se: **Fourierova transformacija** signala $x(t)$, vzorenega s periodo $\Delta t$ je definirana kot vsota:

$$
X_s(f)=\sum_{n=-\infty}^{+\infty}\,x(n\,\Delta t)\,e^{-\textrm{i}\,2\pi\,f\,n\,\Delta t}
$$

in je v frekvenni domeni periodina s frekvenco $1/\Delta t$:

$$
X_s(f)=X_s(f+r/\Delta t), \qquad \textrm{kjer je:}\quad r\in \mathbb{Z}.
$$
:::

### Frekvenno prekrivanje (*aliasing*)

Posledica periodinosti Fourierove transformacije diskretnega signala je, da so frekvenne komponente signla, vije od $f_s/2$, zrcaljene preko meje $f_s/2$ in povzroajo t. i. **frekvenno prekrivanje**. To povzroa anomalije v diskretno vzorenem signalu.

Frekvenco $f_N = f_s / 2$ imenujemo tudi *Nyquistova frekvenca*.

:::{note}
Pojav frekvennega prekrivanja je zanimivo opazovati na primeru digitalnih slik, kjer je neodvisna spremenljivka razdalja (viina, irina slike) in govorimo o *prostorski frekvenci*.

![image-aliasing](./images/07/image_aliasing.png)
:::

```python
X = np.fft.rfft(x) / len(x)
X[1:] *= 2
freq = np.fft.rfftfreq(len(x), dt)
```

### Kvantizacija

Kvantizacija je postopek diskretizacije *odvisne spremenljivke* zveznega procesa $x(t)$. 

Glavni parameter asovnega vzorenja je **tevilo diskretnih nivojev**, ki jih odvisna spremenljivka pri procesu kvantizacije lahko zavzame, ki doloa **dinamini razpon** (globino) procesa. Obiajno je podano s **tevilom bitov** $b$ AD pretvornika. Prav tako je pri tem pomembno obmoje kvanzizacije $A = x_{\text{max}} - x_{\text{min}}$, ki doloa interval monih vrednosti neodvisne spremenljivke $x \in [x_{\text{min}}, x_{\text{max}}]$.

tevilo diskretnih vrednosti kvantiziranega signala je:

$$N = 2^b$$

Napako kvantizacije lahko modeliramo kot nakljuni um normalne porazdelitve s standardno deviacijo:

$$\sigma_e = \frac{A/2^b}{\sqrt{12}}$$

*Dinamini razpon* meritve je doloen z razmerjem med koristnim signalom in umom (napako kvantizacije), *SNR* (*Signal-Nosie Ratio*):

$$SNR = 10 \, \log_{10} \, \frac{\sigma_x^2}{\sigma_e^2} = 20 \, \log_{10} \, \frac{\sigma_x}{\sigma_e} \approx 6\,b - 1.25 dB $$

(ob predpostavki $\sigma_x = A/4$, da ne doseemo mej dovoljenga intervala neodvisne spremenljivke).

```python
adc_range = [0, 5]
A = np.max(adc_range) - np.min(adc_range) 
b = 10
std_e = (A/2**b) / np.sqrt(12)
std_x = A/4
```

```python
SN = 10 * np.log10(std_x**2 / std_e**2)
SN
```

```python
6*b - 1.25
```

```python
def get_quantized(x, bits=6, adc_range=(-20, 20)):
    """
    Kvantizacija signala `x` z izbranimi parametri (s predavanj).
    """
    x2 = x.copy()
    lo, hi = adc_range
    x2[x<=lo] = lo
    x2[x>=hi] = hi
    delta = (hi - lo) / (2**(bits)-1)
    qnt = lo + delta*np.floor((x2 - lo) / delta)
    return qnt
```

Kvantiziran signal

```python
x_kvantiziran = get_quantized(x, b, adc_range)
```

Signal s umom

```python
x_um = x + np.random.randn(len(x))*std_e
```

```python
plt.plot(t, x, label='x(t)')
plt.plot(t, x_um, label='um')
plt.plot(t, x_kvantiziran, '.-', label='po kvantizaciji')
plt.xlabel('t [s]')
plt.ylabel('x(t)')
plt.legend();
```

```python
X_k = np.fft.rfft(x_kvantiziran) / len(x_kvantiziran) 
X_k[1:] *= 2

X_um = np.fft.rfft(x_um) / len(x_um) 
X_um[1:] *= 2
```
