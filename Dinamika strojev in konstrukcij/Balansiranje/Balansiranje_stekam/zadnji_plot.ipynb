```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')  # Use the default style from your environment
plt.rc('font', family='serif', size=10)
plt.rcParams['text.usetex'] = False
```

```python
a = [0, 1, 2]
k = [9.715, 0.615, 0.051]
fig, ax = plt.subplots(figsize=(8, 1.5))
ax.semilogy(a, k, '.-')
ax.set_xticks(a)
ax.set_xlabel('Korak balansiranja')
ax.set_ylabel('Amplituda [N]')
ax.set_ylim(0.01, 100)

for i, txt in enumerate(k):
    ax.annotate(f'{txt:.3f}', (a[i], k[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.savefig('Amp.jpg', dpi=300, bbox_inches='tight')
plt.savefig('C:/Users/marko/Desktop/Adams/slike/Amp.jpg', dpi=300, bbox_inches='tight')
```
