```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1;
k = 1000; #N/m

m1 = m; 
m2 = 2*m;
k1 = 2*k;
k2 = 2*k;
```

```python
M = np.array([[m1, 0], [0,m2]])
K = np.array([[k1+k2, -k2], [-k2, k2]])
A = np.linalg.inv(M)@K
print("matrix M:")
print(M)
print("matrix K:")
print(K)
print("matrix A:")
print(A)
```

```python
#eigenvalues and eigenvectors
w, v = np.linalg.eig(A);
```

```python
w = np.sqrt(w)
print(w)
# plt.plot(w, 'o')
#plot v
# plt.show()
plt.plot(v[:,0], 'o')
# plt.plot(v[:,1], 'o')
```
