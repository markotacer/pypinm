```python
import math
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

def calculate_point_B_speed(piston_velocity, crank_radius, connecting_rod_length, angle):
    # Calculate angular velocity of the crank
    angular_velocity = piston_velocity / (crank_radius * math.cos(math.radians(angle)))
    
    # Calculate speed of point B
    point_B_speed = angular_velocity * crank_radius
    
    return point_B_speed

# Input values
piston_velocity = 2.5  # in m/s
crank_radius = 0.1  # in meters
connecting_rod_length = 0.2  # in meters
angle = 30  # in degrees

# Calculate speed of point B symbolically
point_B_speed_symbolic = calculate_point_B_speed('v_piston', 'r_crank', 'connecting_rod_length', 'angle')

# Calculate speed of point B numerically
point_B_speed_numeric = calculate_point_B_speed(piston_velocity, crank_radius, connecting_rod_length, angle)

# Plotting
fig, ax = plt.subplots()

# Plotting crank radius
ax.plot([0, crank_radius * math.cos(math.radians(angle))], [0, crank_radius * math.sin(math.radians(angle))], 'b-', label='Crank Radius')

# Plotting connecting rod
connecting_rod_end_x = crank_radius * math.cos(math.radians(angle)) + connecting_rod_length * math.cos(math.radians(angle + 90))
connecting_rod_end_y = crank_radius * math.sin(math.radians(angle)) + connecting_rod_length * math.sin(math.radians(angle + 90))
ax.plot([crank_radius * math.cos(math.radians(angle)), connecting_rod_end_x], [crank_radius * math.sin(math.radians(angle)), connecting_rod_end_y], 'r-', label='Connecting Rod')

# Plotting piston as a rectangle
piston_width = 0.03  # Width of the piston rectangle
piston_height = 0.08  # Height of the piston rectangle
piston_center_x = crank_radius * math.cos(math.radians(angle)) + (connecting_rod_length/2) * math.cos(math.radians(angle + 90))
piston_center_y = crank_radius * math.sin(math.radians(angle)) + (connecting_rod_length/2) * math.sin(math.radians(angle + 90))
piston_angle = math.degrees(math.atan2(connecting_rod_end_y - crank_radius * math.sin(math.radians(angle)), connecting_rod_end_x - crank_radius * math.cos(math.radians(angle))))
piston_rect = Rectangle((piston_center_x - piston_width/2, piston_center_y - piston_height/2), piston_width, piston_height, angle=piston_angle, edgecolor='g', facecolor='g')
ax.add_patch(piston_rect)

# Configure plot
ax.set_aspect('equal', 'box')
ax.set_xlim(-0.1, max(crank_radius, connecting_rod_end_x) + 0.1)
ax.set_ylim(-0.1, max(crank_radius, connecting_rod_end_y) + 0.1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

plt.show()

# Print results
print("Speed of point B (symbolic):", point_B_speed_symbolic)
print("Speed of point B (numerical):", point_B_speed_numeric)
```
