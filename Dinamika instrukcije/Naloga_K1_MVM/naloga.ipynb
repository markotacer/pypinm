```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
a = 0.8 # m
l = 1.2 # m
mu = 18 # kg/m
```

```python
# Palice obravnavamo posebej, 1 - spodnja, 2 - vertikalna in 3 - zgornja
m1 = mu * l
m2 = mu * a
m3 = mu * 2 * a
# Palica1
dx1 = l/2
dy1 = 0
dz1 = l/2
Jxx1 = 1/12 * m1 * l**2 + m1 * (dx1)**2
Jyy1 = 0
Jzz1 = 1/12 * m1 * l**2 + m1 * (dz1)**2
# Palica 2
dx2 = np.sqrt((a/2)**2 + (l)**2)
dy2 = a/2
dz3 = l
Jxx2 = 1/12 * m2 * a**2 + m2 * (dx2)**2
Jyy2 = 1/12 * m2 * a**2 + m2 * (dy2)**2
Jzz2 = 0 + m2 * (dz3)**2
# Palica 3
dx3 = np.sqrt((a)**2 + (l)**2)
dy3 = a
dz3 = l
Jxx3 = 0 + m3 * (dx3)**2
Jyy3 = 1/12 * m3 * (2*a)**2 + m3 * (dy3)**2
Jzz3 = 1/12 * m3 * (2*a)**2 + m3 * (dz3)**2

Jxx = Jxx1 + Jxx2 + Jxx3
Jyy = Jyy1 + Jyy2 + Jyy3
Jzz = Jzz1 + Jzz2 + Jzz3

print(f'Jxx: {Jxx:.2f} kgm^2')
print(f'Jxx1: {Jxx1:.2f} kgm^2')
print(f'Jxx2: {Jxx2:.2f} kgm^2')
print(f'Jxx3: {Jxx3:.2f} kgm^2')
print()
print(f'Jyy: {Jyy:.2f} kgm^2')
print(f'Jyy1: {Jyy1:.2f} kgm^2')
print(f'Jyy2: {Jyy2:.2f} kgm^2')
print(f'Jyy3: {Jyy3:.2f} kgm^2')
print()
print(f'Jzz: {Jzz:.2f} kgm^2')
print(f'Jzz1: {Jzz1:.2f} kgm^2')
print(f'Jzz2: {Jzz2:.2f} kgm^2')
print(f'Jzz3: {Jzz3:.2f} kgm^2')
```

```python
# e deviacijski momenti in momenti inercije v teiu
Jxy1 = 0
Jxz1 = 0 + m1 * dx1 * dz1
Jyz1 = 0

Jxy2 = 0 + m2 * dx2 * dy2
Jxz2 = 0 + m2 * dx2 * dz3
Jyz2 = 0 + m2 * dy2 * dz3

Jxy3 = 0 + m3 * dx3 * dy3
Jxz3 = 0 + m3 * dx3 * dz3
Jyz3 = 0 + m3 * dy3 * dz3


Jxy = Jxy1 + Jxy2 + Jxy3
Jxz = Jxz1 + Jxz2 + Jxz3
Jyz = Jyz1 + Jyz2 + Jyz3

print(f'Jxy: {Jxy:.2f} kgm^2')
print(f'Jxz: {Jxz:.2f} kgm^2')
print(f'Jyz: {Jyz:.2f} kgm^2')
```

```python
rx = 0.0
ry = l
rz = a

R = np.sqrt((rx)**2 + (ry)**2 + (rz)**2)
lx = rx/R
ly = ry/R
lz = rz/R
l = [lx, ly, lz]
print(f'lx: {lx:.2f}')
print(f'ly: {ly:.2f}')
print(f'lz: {lz:.2f}')
```

```python
def Jp(Jxx, Jyy, Jzz, Jxy, Jxz, Jyz, l):
    Jp = l[0]**2 * Jxx + l[1]**2 * Jyy + l[2]**2 * Jzz + 2 * l[0] * l[1] * Jxy \
    - 2 * l[0] * l[2] * Jxz - 2 * l[1] * l[2] * Jyz
    return Jp

JP = Jp(Jxx, Jyy, Jzz, Jxy, Jxz, Jyz, l)
print(f'JP: {JP:.2f} kgm^2')
```
