## Skarpa 2

```python
g = 0.3 # globina zgornje police
l1 = 1.8
l2 = l1 + 2.6

h1 = 0.6
h2 = 1.2
h3 = 1.2

A1 = (h1+h2)/2 * l1
A2 = (h2+h3)/2 * (l2-l1)
A = A1+A2

L1 = sp.spatial.distance.euclidean((0, h1), (l1, h2))
L2 = sp.spatial.distance.euclidean((l1, h2), (l2, h3))

x1 = np.array([[0, l1, l2]])
y1 = np.array([[h1, h2, h3]])
L = L1 + L2
A4 = L1 * g
A02 = A + A4

plt.plot(x1, y1, '-o', c = 'C0')
plt.annotate(f'{A1:.2f} m2', (1, 0.5))
plt.annotate(f'{A2:.2f} m2', (3, 0.5))
plt.fill(x1, y1)
plt.ylim(0, 1.4)
plt.xlim(0, 5)
plt.show()
print(f'Dolžina na vrhu = {L1:.3f} m')
print(f'Površina na vrhu = {A4:.3f} m2')
print(f'Skupna površina = {(A4 + A):.3f} m2')
```
