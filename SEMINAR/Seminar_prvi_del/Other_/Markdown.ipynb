### Primer naloge

Kolo se kotali po podlagi definirani s $y(t)$, in podpira maso $m$ s vzmetjo togosti $k$.

```python
g = 9.81 #ms-2
v = 50 #m/s
x_st = 100 #mm
T = np.arange(0, 10.01, 0.01)
t = np.array(T)
z = v * t
y0 = 20 #mm
y = y0 * np.sin((np.pi * z)/10)
```

Iščemo amplitudo nihanja mase

Lastno frekvenco sistema masa vzmet določimo:
$$\omega_0 = {(\frac{g}{x_{st}})}^{1/2}$$

```python
omega0 = np.sqrt(g/(x_st/1000))
print(f' {omega0:.2f} rad/s')
```

$$z = v  t$$
$$y(t) = y_0 \cdot sin(5 \pi t )$$

Upoštevamo enačbo:
$$\frac{X}{Y} = \frac{\sqrt{1+(2 \delta \frac{\omega}{\omega_0})^2}}
{\sqrt{(1 - (\frac{\omega}{\omega_0})^2)^2+(2 \delta \frac{\omega}{\omega_0})^2}}$$

```python
omega = 5*np.pi
delta = 0
```

```python
def x(o, o0, d = 0, Y = 1):
    
    A0 = np.sqrt(1 + (2 * d * o/o0)^2)
    B0 = np.sqrt((1 - (o/o0)^2)^2 + (2 * d * o/o0)^2)
    return Y * A0/B0
```

```python
x1 = x(o = omega, o0 = omega0, d = delta, Y = y0)
x1
```
